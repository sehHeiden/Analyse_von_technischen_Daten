```python
from pathlib import Path

import pandas as pd
from matplotlib import pyplot as plt
```

```python
# öffnen der `.csv`-Dateien aus dem aktuellen Ordner (".")
eis_dfs = [pd.read_csv(x, index_col=0, skiprows=1) for x in Path().iterdir() if x.suffix == ".csv"]
eis_dfs[0]
```

```python
for eis_df in eis_dfs:
    plt.plot(eis_df["Re/Ohm"], -1 * eis_df["Im/Ohm"])
```

```python
max_im = [max(-1 * x["Im/Ohm"]) for x in eis_dfs]
cycles = [int(x.stem.split("_")[0]) for x in Path().iterdir() if x.suffix == ".csv"]
plt.plot(cycles, max_im, "r+")
```
